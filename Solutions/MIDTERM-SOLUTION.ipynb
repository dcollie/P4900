### Q6 - Add a column to your data!
- Call this column: 'Mass'
- If Spectral Class is M, K, G, F, A the value of the column has to be 'Low-Mass'
- If Spectral Class is O, or B the value of the column has to be 'High-Mass'


```python
def Mass(x):
    if x in 'MKGFA':
        return 'Low-Mass'
    return 'High-Mass'

df['MASS']=df['Spectral Class'].apply(Mass)
```

# Part IV
Perform Linear Regression on your data utilizng: 'Temperature (K)', 'Radius(R/Ro)', 'Absolute magnitude(Mv)' 

to predict 'Luminosity(L/Lo)'
- Evaluate your model
- Is it a good fit?

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df.columns
```

```python
X=df[['Temperature (K)', 'Radius(R/Ro)', 'Absolute magnitude(Mv)']]
y=df['Luminosity(L/Lo)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
pd.DataFrame(data=np.append(lm.intercept_, lm.coef_),
             index=np.append('intercept',X.columns),
             columns=['Coefficient']
            )
```

```python
pred=lm.predict(X_test)
```

```python
sns.scatterplot(y=y_test, x=pred)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, pred))
```

```python
(y_test.min(), y_test.max(), y_test.mean())
```

```python
[np.sqrt(mean_squared_error(y_test, pred))/i for i in [y_test.min(), y_test.max(), y_test.mean()]]
```
