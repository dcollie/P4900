Potability

Water potability refers to the suitability of water for consumption by humans. A potable water supply is one that is free from harmful contaminants, pathogens, and other substances that may cause health problems.

There are several criteria that must be met for water to be considered potable. These include:

- Microbial safety: Water must be free from harmful microorganisms, such as bacteria, viruses, and parasites that can cause illness.
- Chemical safety: Water must be free from harmful chemical contaminants, such as heavy metals, pesticides, and industrial chemicals that can cause health problems.
- Physical safety: Water must be free from physical contaminants, such as sediment, particulate matter, and other visible materials.
- Taste and odor: Water should be free from unpleasant taste and odor that may make it less desirable for consumption.

To ensure water potability, drinking water is typically treated through various processes that can remove or reduce harmful contaminants and pathogens. These processes can include filtration, disinfection, and chemical treatment.

Water quality standards and regulations vary by location and can be set by national or local authorities. In the United States, the Environmental Protection Agency (EPA) sets standards for drinking water quality under the Safe Drinking Water Act. These standards define maximum contaminant levels for various contaminants and parameters, and are regularly reviewed and updated as new information becomes available.

Overall, ensuring water potability is essential for protecting public health and ensuring access to safe and clean drinking water

pH

Description: pH is a measure of the acidity or alkalinity of a solution, including water. The pH scale ranges from 0 to 14, where a pH of 7 is considered neutral, below 7 is acidic, and above 7 is alkaline.

Measurement: pH

Potabiltiy criterion: Chemical Safety, Taste and Odor

Safe limits: 6.5 to 8.5 (World Health Organization (WHO))

Below limit: 
- can indicate the presence of dissolved metals such as lead, copper, and zinc, which can cause health problems
- can affect taste

Above limit:  
- can indicate the presence of minerals such as calcium and magnesium, which can contribute to hard water
- can affect taste

Treability: neutralization with chemicals, aeration, or revese osmosis

Other concerns: acidic water can also cause corosion to pipes and other plumbing fixtures. pH levels can also affect water treatment processes. For example, some water treatment processes, such as disinfection with chlorine, work best at a specific pH range.

Hardness

Definition: Hardness is a measure of the concentration of dissolved minerals, primarily calcium and magnesium, in water. 

Measurement: parts per million

Potabiltiy criterion: Taste and Odor

Safe limits: 200 mg/L (or parts per million)

Above limit: 
- can affect taste, and texture and appearance

Treatabitlity: Water with a hardness level above the limit may be treated through various methods such as ion exchange, reverse osmosis, or lime softening to reduce the concentration of dissolved minerals and improve its quality

Other concerns: These minerals can accumulate in pipes and other plumbing fixtures, causing scaling and other problems. Hard water can also make soap less effective, leading to decreased lathering and cleaning ability, and can cause mineral buildup in appliances such as dishwashers and water heaters, reducing their efficiency and lifespan.


Solids

Description: Solids in water can include both suspended solids, which are particles that are floating in the water, and dissolved solids, which are particles that are fully dissolved in the water, measured as the total dissolved solids (TDS), which includes dissolved minerals such as calcium and magnesium, as well as other dissolved substances like salts and organic compounds.

Meaurement: parts per million

Potabiltiy criterion: Physical Saftey, Chemical Safety, Taste and Odor

Safe limits: maximum level of 500 mg/L for TDS (World Health Organization (WHO))

Above limit: It can lead to unpleasant taste, odor, and color, and can also cause health problems in some cases.  Dissolved minerals such as sodium, magnesium, and sulfate can cause gastrointestinal problems or contribute to hypertension. 

Treatability: reverse osmosis or distillation can be used to remove dissolved solids

Other concerns: can reduce the effectiveness of disinfection processes, and can cause scaling and buildup in pipes and other plumbing fixtures.

Chloramines

Description: Chloramines are a type of disinfectant that is commonly used in drinking water treatment to kill harmful bacteria and other pathogens. While chloramines are effective at disinfection, they can also have negative effects on water quality and potability.

Potabiltiy criterion: Chemical Safety, Taste and Odor

Safe limits: maximum contaminant level (MCL) of 4 parts per million (ppm) or milligrams per liter (mg/L) for total chlorine which includes both free chlorine and chloramines.

Above limit: 
- can also affect the taste, odor, and appearance of drinking water,
- can react with organic matter in the water, forming potentially harmful byproducts like trihalomethanes (THMs) and haloacetic acids (HAAs). These byproducts can pose health risks, particularly if they are present in high concentrations over long periods of time.

Treatability: Water treatment processes such as activated carbon filtration or reverse osmosis can be used to remove chloramines and their byproducts.

Sulfate

Description: Sulfate is a naturally occurring mineral that is commonly found in drinking water sources. 

Potabiltiy criterion: Microbial Saftey, Chemical Safety, Taste and Odor

Safe limits: secondary maximum contaminant level (SMCL) of 250 milligrams per liter (mg/L) or parts per million (ppm) for sulfate in drinking water Environmental Protection Agency (EPA)

Above limit: 
- can react with other substances in the water, such as metals, to form sulfates or sulfides that can cause taste and odor problems
- can have a laxative effect, leading to diarrhea and dehydration in some cases,

Other concerns: sulfate-reducing bacteria thrive in plumbing systems that have low oxygen levels, such as in hot water heaters or stagnant water and can produce hydrogen sulfide gas, which has a distinctive and unpleasant rotten egg odor.

Treatability: reverse osmosis, ion exchange, or distillation.


Conductivity

Description: Conductivity is a measure of the ability of water to conduct an electrical current, which is influenced by the concentration of dissolved ions in the water.

Potabiltiy criterion: Chemical Safety, Taste and Odor

Safe limits: secondary maximum contaminant level (SMCL) of 500 microsiemens per centimeter (μS/cm) (Environmental Protection Agency (EPA))

Above limit: 
- can have a salty or brackish taste, 
- can indicate the presence of dissolved minerals, such as sodium, chloride, and sulfate which can have secondary effects, 
- can also indicate the presence of other contaminants in drinking water, such as heavy metals or organic compounds which can have negative health effects, even at low concentrations.

Treatability: reverse osmosis or distillation.

Other concerns: can indicate dissolved minerals which can contribute to corrosion of plumbing systems and appliances.

Organic Carbon

Description: Organic carbon is a measure of the amount of carbon-based compounds present in water, which can be naturally occurring or a result of human activities.

Potabiltiy criterion: Mibcrobial Safety, Chemical Safety, Taste and Odor

Safe limits: maximum contaminant level (MCL) of 10 milligrams per liter (mg/L) for total organic carbon (TOC)

Above limit: 
- reacts with chlorine to form harmful disinfection byproducts (DBPs), such as trihalomethanes (THMs). THMs have been linked to an increased risk of cancer and other negative health effects.
- can provide a food source for bacteria and other microorganisms, which can lead to bacterial growth and contamination of drinking water.

Treatabiltiy: activated carbon filtration, ozone treatment, or ultraviolet (UV) disinfection

Trihalomethanes

Description: Trihalomethanes (THMs) are disinfection byproducts that can form in drinking water when chlorine or other disinfectants react with organic matter, such as humic and fulvic acids, present in the water.

Potabiltiy criterion: 

Safe limits: maximum contaminant level (MCL) of 80 parts per billion (ppb) for total THMs (TTHMs) (Environmental Protection Agency (EPA))

Above limit: 
- have been linked to an increased risk of cancer and other negative health effects.
- can affect the taste and odor of drinking water

Treatability: carbon filtration or ozonation can be used to remove organic matter and reduce THM formation

Turbidity

Description: Turbidity is a measure of the cloudiness or haziness of water caused by the presence of suspended particles, such as sediment, algae, or other microorganisms.

Potabiltiy criterion: Microbial Safety, Physical Safety, Taste and Odor

Safe limits: maximum contaminant level (MCL) of 1.0 nephelometric turbidity units (Health Canada)

Above limit: 
- can make it difficult to disinfect water effectively, as particles can shield bacteria and other microorganisms from the disinfectant.
- can increase the risk of waterborne illness and contamination
- can also affect the taste and odor of drinking water

Treatability: filtration, coagulation, and sedimentation

```python
# acceptable limits
limits_lower = [6.5,0,0,0,0,0,0,0,0]
limits_upper = [8.5,200,500,4,250,500,10,100,1]
```

```python
wp1 = wp
X1 = wp1[cols]
s = pd.Series(data=wp1['Potability'])
wp1['Potable_Dummies'] = pd.get_dummies(s, drop_first = True, dtype = int)
Y1 = wp1['Potable_Dummies']
X1_train, X1_new, y1_train, y1_new = train_test_split(X1,Y1, test_size=0.30, random_state=42)
X1_validation, X1_test, y1_validation, y1_test = train_test_split(X1_new,y1_new, test_size=0.50, random_state=42)
scal1 = StandardScaler()
scld_X1_train = scal1.fit_transform(X1_train)
scld_X1_validation = scal1.transform(X1_validation)
scld_X1_test = scal1.transform(X1_test)
scaled_X1_train = pd.DataFrame(scld_X1_train, columns=X1.columns)
scaled_X1_validation = pd.DataFrame(scld_X1_validation, columns=X1.columns)
scaled_X1_test = pd.DataFrame(scld_X1_test, columns=X1.columns)
```

```python
ANN = Sequential()
ANN.add(Dense(18, activation='relu'))
ANN.add(Dropout(0.5))
ANN.add(Dense(10, activation='relu'))
ANN.add(Dropout(0.5))
ANN.add(Dense(10, activation='relu'))
ANN.add(Dropout(0.5))
ANN.add(Dense(1, activation='sigmoid'))
ANN.compile(optimizer='adam', loss='binary_crossentropy', )
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
ANN.fit(x=scld_X1_train, y=y1_train, 
          epochs = 600, 
          verbose=1, 
          batch_size = 31,
          callbacks= [early_stop], 
          validation_data=(scld_X1_validation, y1_validation))
```

```python
loss_df1 = pd.DataFrame(ANN.history.history)
loss_df1.plot()
```

```python
ANN_pred_y = np.round(ANN.predict(scld_X1_test))
print(confusion_matrix(y1_test, ANN_pred_y))
print(classification_report(y1_test, ANN_pred_y))
```
