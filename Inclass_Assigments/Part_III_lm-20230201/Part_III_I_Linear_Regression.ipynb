# Act 3 - Linear Regression Model  & Prediction
Let's say in real life we cannot directly measure the 'stator_tooth' temperature. However, we have access to the temperature of 'coolant', 'stator_winding', & 'stator_yoke'. Can you fit a model to these to predictthe temperature of 'strator_tooth'?

#### 3.1 Fit the model
- Create X & Y
- Split your data into training and set
- Import your linear regression model
- Fit your model to your data
- Check the coefficients 
- **What does each coefficient tell you?**
- What is the intercept?

```python
cat.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cat[['coolant','stator_winding','stator_yoke']], 
                                 cat['stator_tooth'], test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

Each coefficient is related to which of your parameters?

```python
lm.coef_
```

```python
lm.intercept_
```

#### 3.2 Predict y$_{test}$ and evaluate your pmodel!
- Make the prediction
- A scatter plot might help to visualize your Fit. Make sure to label your axis
- Use RMSE to evaluate your fit!
- Is this a good fit? Why? Check your mean, min, & max!

```python
y_pred = lm.predict(X_test)
sns.scatterplot(x=y_pred, y=y_test)
plt.xlabel('Predictions')
plt.ylabel('Y$_{test}$')
```

```python
from sklearn import metrics
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
[y_test.mean(), y_test.min(), y_test.max()]
```

```python
X = cat[['u_q', 'coolant', 'stator_winding', 'u_d','motor_speed',
         'i_d', 'i_q', 'pm', 'stator_yoke', 'ambient', 'torque']]
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, 
    cat['stator_tooth'], test_size=0.33, random_state=42)
```

```python
lm2=LinearRegression()
```

```python
lm2.fit(X_train2, y_train2)
```

```python
lm2.coef_
```

```python
lm2.intercept_
```

```python
new = pd.DataFrame(data=lm2.coef_,
             index = X.columns,
            columns = ['Coefficient'])
new
```

```python
new.sort_values(by='Coefficient', ascending=False)
```

#### 4.2 Predict y$_{test}$ and evaluate your model!

```python
y_pred2 = lm2.predict(X_test2)
sns.scatterplot(x=y_pred2, y=y_test2)
plt.xlabel('Predictions')
plt.ylabel('Y$_{test}$')
```

```python
np.sqrt(metrics.mean_squared_error(y_test2, y_pred2))
```
