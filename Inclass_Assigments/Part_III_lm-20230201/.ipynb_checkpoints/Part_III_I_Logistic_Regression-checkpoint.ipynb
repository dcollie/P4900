#### 2.3 Let's remind ourselves of the behavior of our data points
    - Could you use a "single" comprehension list to print the following list from your dataframe.columns?
    
**Hint: approach step by step**

Comprehension List: https://www.w3schools.com/python/python_lists_comprehension.asp
    

```python
cols1 = []
for n in cat.columns:
    if 'ID' not in n:
        if 'DES' not in n:
            if 'ULIM' not in n:
                if "_TOT" not in n:
                    if "_ADD" not in n:
                        if "DF" not in n:
                            if "FSC" not in n:
                                if "DMASS" not in n:
                                    if "_TYPE" not in n:
                                        if "_LARS" not in n:
                                            if "_DIST" not in n:
                                                if "DIST_" not in n:
                                                    if "DTEMP" not in n:
                                                        if "LR" not in n:
                                                            if "LAM" not in n:
                                                                  cols1.append(n)
print(cols1)
```

#### 2.4 Call the list above 'cols'

#### 2.5 From 'cols' we only need ['F70', 'F160', 'F250', 'F350', 'F500', 'F21', 'F22', 'F24', 'F870', 'F1100']. Delete the rest!  


```python
cols = []
for n in cols1:
    if n.startswith('F') == True:
        cols.append(n)
```

```python
cols
```

#### 2.6 From your data delete all the columns with EVOL_FLAG == 0
**Use this dataset for your train_test_split in the next section**

```python
cat.drop(cat[cat['EVOL_FLAG']==0].index, inplace = True)
```

```python
cat
```

# Act 3 - Use sklearn and Logistic Regression to train your data!
- Use cols for your Design Matrix (X)
- Your y column should be the 'EVOL_FLAG'
- Check your Coefficients and Intercept --> What do they represent?
- perform the prediction
- Create a scatter plot of $y_{test}$ vs prediction, is this representitive?
- Create confusion matrix and classification report

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = cat[cols]
y = cat['EVOL_FLAG']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X
```

```python
y
```

```python
lg=LogisticRegression(max_iter=10000)
```

```python
lg.fit(X_train, y_train)
```

```python
lg.coef_
```

```python
pd.DataFrame(data=lg.coef_.reshape(10,1), index = X.columns, columns = ['Coefficients'])
```

```python
lg.intercept_
```

```python
pred_y = lg.predict(X_test)
```

```python
sns.scatterplot(x=pred_y, y=y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix(y_test, pred_y)
```

```python
print(classification_report(y_test, pred_y))
```
