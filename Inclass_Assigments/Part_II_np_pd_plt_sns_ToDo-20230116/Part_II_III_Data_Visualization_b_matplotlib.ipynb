# Act 2 - Create the following plots

#### 2.1 Use object oriented method. Define 15 points from (0,10). Plot $\sqrt x$

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,.8,.8])

x = np.linspace(0,10,15)

ax.plot(x,x**0.5)
```

#### 2.2. Label your axes and add title

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,.8,.8])
ax.set_xlabel('x axis')
ax.set_ylabel('$\sqrt{x}$')
ax.set_title('Root Squared Plot')
ax.plot(x,x**0.5)
```

#### 2.3. Add a smaller second plot of $x^2$ on the right side of this and have appropriate labels in.

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0.1,0.1,.8,.8])
ax2 = fig1.add_axes([.55,.2,.3,.4])

ax1.set_xlabel('x axis')
ax1.set_ylabel('$\sqrt{x}$')
ax1.set_title('Root Squared Plot')
ax1.plot(x,x**0.5)

ax2.set_xlabel('x axis')
ax2.set_ylabel('x$^2$')
ax2.set_title('Squared Plot')
ax2.plot(x,x**2, c='r', ls= '--')
```

#### 2.4. Can you update it to look like this?

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0.1,0.1,.8,.8])
ax2 = fig1.add_axes([.55,.2,.3,.4])

ax1.set_xlabel('x axis')
ax1.set_ylabel('$\sqrt{x}$')
ax1.set_title('Root Squared Plot')
ax1.plot(x,x**0.5, c='c', marker='X', ls=':', markeredgecolor='b', markeredgewidth = 2, markersize = 10, lw = 5, markerfacecolor = 'y')

ax2.set_xlabel('x axis')
ax2.set_ylabel('x$^2$')
ax2.set_title('Squared Plot')
ax2.plot(x,x**2, c='r', ls= '--')
```

# Act 3 - Subplots

#### Create the following plots

```python
fig, ax = plt.subplots(nrows=1, ncols = 2,figsize=(10,4))

ax[0].plot(x,x**2, c='r', lw = 5)
ax[1].plot(x,x**0.5)
ax[0].set_xlabel('This title is the same as the one next to it')
ax[1].set_xlabel('This title is the same as the one next to it')
ax[0].set_ylabel('x$^2$')
ax[1].set_ylabel('$\sqrt{x}$')
```

#### 3.2. Add $20\cdot\sqrt{x}$ to the plot on the left, Make sure to add the legends in

```python
fig, ax = plt.subplots(nrows=1, ncols = 2,figsize=(10,4))

ax[0].plot(x,x**2, c='r', lw = 8, alpha = .5, label = 'x$^2$')
ax[0].plot(x,20*x**0.5, c='k', lw = 8, alpha = .5, label = '20$\cdot\sqrt{x}$')
ax[1].plot(x,x**0.5)

for n in ax:
    n.set_xlabel('This title is the same as the one next to it')
ax[0].set_ylabel('x$^2$ and 20$\cdot\sqrt{x}$')
ax[1].set_ylabel('$\sqrt{x}$')
ax[0].legend()
```
