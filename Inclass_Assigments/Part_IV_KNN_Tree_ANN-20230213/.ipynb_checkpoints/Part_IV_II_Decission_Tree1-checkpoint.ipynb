# Act 2 - Decision Tree and Random Forest to Predict Star Type
**Note: This example is for educational purposes only**


#### 2.1.1 Decisison Tree
- X should contain ['Temperature (K)', 'Luminosity(L/Lo)', 'Radius(R/Ro)','Absolute magnitude(Mv)']
- y is ['Star Type']
- perform train test split
- create your decision tree model
- fit the model to your training set
- perform the predictions on your test set
- evaluate your model (i.e. confusion matrix and classification_report)

```python
df.columns
```

```python
X = df[['Temperature (K)', 'Luminosity(L/Lo)', 'Radius(R/Ro)','Absolute magnitude(Mv)']]
y = df['Star type']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred = dtree.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

#### 2.1.2 Plot your decission tree
- Look at the number of counts in each of the Star Type for y_train
- plot your model (decision tree)

```python
y_train.value_counts()
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(dtree, feature_names = X.columns, class_names = True, filled = True)
```

#### 2.2.1 Now let's bring the random forest into the game!
- X, y and train test split may remain the same as before
- create your random forest model for 5 estimators
- fit the model to your training set
- perform the predictions on your test set
- evaluate your model (i.e. confusion matrix and classification_report)

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=5)
rf.fit(X_train, y_train)
rf_y_pred = rf.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, rf_y_pred))
print(classification_report(y_test,rf_y_pred))
```

#### 2.2.2 Plot your random forest (only 5 trees)
- plot the tree for each of the estimators in your random forest!

```python
for n in rf.estimators_:
    plt.figure(figsize=(15,15))
    tree.plot_tree(n, feature_names = X.columns, class_names = True, filled = True)
```

# Act 3 - Decision Tree and Random Forest on SF data
#### 3.1.1 Decision Tree
- read your star formation data --> The correct formatting
- X should contain ['F70', 'F160', 'F250', 'F350', 'F500', 'F21', 'F22', 'F24', 'F870', 'F1100']
- y is ['Evol Flag']
- perform train test split
- create your decsion tree model
- fit the model to your training set
- perform the predictions on your test set
- evaluate your model (i.e. confusion matrix and classification_report) 

```python
cat=pd.read_csv('../Data/Star_Formation/HiGAL_clump_catalogue_correct_formatting.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

cols = ['F70', 'F160', 'F250', 'F350', 'F500', 'F21', 'F22', 'F24', 'F870', 'F1100']
X2=cat[cols]
y2=cat['EVOL_FLAG']
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.33, random_state=42)

dtree2=DecisionTreeClassifier()
dtree2.fit(X_train2, y_train2)
y_pred2=dtree2.predict(X_test2)

from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test2, y_pred2))
print(classification_report(y_test2, y_pred2))
```

#### 3.1.2 Plot your decission tree

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(dtree2, feature_names = X2.columns, class_names = True, filled = True)
```

#### 3.2.1 Is the result from the random forest any better? 
- X, y and train test split may remain the same as before
- create your random forest model for 5 estimators
- fit the model to your training set
- perform the predictions on your test set
- evaluate your model (i.e. confusion matrix and classification_report)

```python
from sklearn.ensemble import RandomForestClassifier
rf2 = RandomForestClassifier(n_estimators=5)
rf2.fit(X_train2, y_train2)
rf2_y_pred = rf2.predict(X_test2)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test2, rf2_y_pred))
print(classification_report(y_test2,rf2_y_pred))
```

#### 3.2.2 Plot your random forest (only 5 trees)
- plot the tree for each of the estimators in your random forest!

```python
for n in rf2.estimators_:
    plt.figure(figsize=(15,5))
    tree.plot_tree(n, feature_names = X2.columns, class_names = True, filled = True)
```

#### 3.3 Re-run your code in 3.2.1, however, set n_estimator = 200. Does it improve your result?

```python
rf2_1 = RandomForestClassifier(n_estimators=200)
rf2_1.fit(X_train2, y_train2)
rf2_1_y_pred = rf2_1.predict(X_test2)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test2, rf2_1_y_pred))
print(classification_report(y_test2,rf2_1_y_pred))
```
