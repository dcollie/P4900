### Q6 - Add a column to your data!
- Call this column: 'Mass'
- If Spectral Class is M, K, G, F, A the value of the column has to be 'Low-Mass'
- If Spectral Class is O, or B the value of the column has to be 'High-Mass'


```python
def add_mass(x):
    if x == 'M' or x=='K' or x=='G' or x=='F' or x == 'A':
        designation = 'Low-Mass'
    elif x == 'O' or x=='B':
        designation = 'High_Mass'
    return designation

cat['Mass'] = cat['Spectral Class'].apply(add_mass)
cat
```

# Part IV
Perform Linear Regression on your data utilizng: 'Temperature (K)', 'Radius(R/Ro)', & 'Absolute magnitude(Mv)' 

to predict 'Luminosity(L/Lo)'
- Create a datafram of intercept and coef! 
- Evaluate your model!
- Is it a good fit?

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = cat[['Temperature (K)', 'Radius(R/Ro)', 'Absolute magnitude(Mv)']]
y = cat['Luminosity(L/Lo)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print (lm.intercept_)
```

```python
print (lm.coef_)
```

```python
pred_y = lm.predict(X_test)
```

```python
sns.scatterplot(x=pred_y, y=y_test)
```

```python
df = pd.DataFrame()
df['Predictions'] = pred_y
df['Y_test'] = y_test
df.corr()
```
